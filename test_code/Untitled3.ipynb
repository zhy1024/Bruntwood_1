```python
import matplotlib.pyplot as plt

data_points = {
    'Annual Rent': 100%,
    'Date of Lease': 8/11,
    'Term': 8/11,
    'Customer break date': 8/11,
    'Term commencement date': 6/11,
    'Review date': 6/11,
}

categories = list(data_points.keys())
values = list(data_points.values())

plt.figure(figsize=(4,5))
plt.bar(categories, values, color='skyblue')
# plt.xlabel('Categories')
plt.ylabel('Counts')
plt.title('Bar chart of the correct extraction number of 6 features.')
plt.xticks(rotation=45, ha='right')

# Display the value on top of each bar
for i, v in enumerate(values):
    plt.text(i, v + 0.1, str(v), ha='center', va='bottom', fontsize=10)

plt.ylim(0, 12)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

data_points = {
    'Annual Rent': 1.0,
    'Date of Lease': 8/11,
    'Term': 8/11,
    'Term commencement date': 8/11,
    'Customer break date': 7/11,
    'Review date': 6/11,
}

categories = list(data_points.keys())
values = list(data_points.values())

plt.figure(figsize=(5,5))
bars = plt.bar(categories, values, color='skyblue')
# plt.xlabel('Categories')
plt.ylabel('Percentage')
plt.title('Bar chart of the accuracy of 6 features in 11 lease files.')
plt.xticks(rotation=45, ha='right')

# Display the value on top of each bar
for i, v in enumerate(values):
    bars[i].set_height(v)
    plt.text(i, v + 0.02, f'{v:.2%}' if v != 1.0 else '100%', ha='center', va='bottom', fontsize=10)

plt.ylim(0, 1.2)  # Set y-axis limit to 0 and 1.2

# Format y-axis ticks and labels as percentages using FuncFormatter
def percent_formatter(x, pos):
    return f'{x:.0%}'

plt.gca().yaxis.set_major_formatter(FuncFormatter(percent_formatter))

plt.tight_layout()
plt.show()
```
